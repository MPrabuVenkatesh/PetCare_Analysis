```python
import pandas as pd
pd.set_option('display.max_columns', 200)

# load
breeds = pd.read_csv('cleaned_dog_breeds.csv')
descriptions = pd.read_csv('cleaned_dog_descriptions.csv')
supplies = pd.read_csv('cleaned_pet_supplies.csv')

# quick shapes / columns
print("breeds:", breeds.shape)
print("descriptions:", descriptions.shape)
print("supplies:", supplies.shape)

print("\nbreeds columns:\n", list(breeds.columns))
print("\ndescriptions columns:\n", list(descriptions.columns))
print("\nsupplies columns:\n", list(supplies.columns))

# quick head
display(breeds.head())
display(descriptions.head())
display(supplies.head())
```

```python
# parse dates
descriptions['posted'] = pd.to_datetime(descriptions['posted'], errors='coerce')
descriptions['accessed'] = pd.to_datetime(descriptions['accessed'], errors='coerce')

# booleans
for c in ['env_children','env_dogs','env_cats']:
    if c in descriptions.columns:
        descriptions[c] = descriptions[c].astype('boolean')

# clean product sold-text into a numeric parsed column (optional)
supplies['tradeamount_parsed'] = supplies['tradeamount'].astype(str).str.replace(r'[^0-9]','', regex=True)
supplies['tradeamount_parsed'] = pd.to_numeric(supplies['tradeamount_parsed'], errors='coerce')

# unify pet_type capitalization
supplies['pet_type'] = supplies['pet_type'].astype(str).str.title()

# quick checks
print(supplies[['tradeamount','tradeamount_parsed','sold_count','pet_type']].head(10))
```

```python
# Merge dog_descriptions (fact) with dog_breeds (dimension) on breed_clean
merged = descriptions.merge(
    breeds,
    on='breed_clean',
    how='left',
    suffixes=('_desc','_dim')
)

print("merged rows:", merged.shape)
print("merged columns:", merged.columns.tolist()[:20])  # check first 20 cols

# Show sample
display(merged[['id','breed_clean','breed_dim','size_category_dim']].head())
```

```python
# --- Clean longevity__dim into numeric values ---
def parse_longevity(val):
    if pd.isna(val):
        return None
    val = str(val).strip()
    # Handle ranges like "12-15"
    if "-" in val:
        try:
            lo, hi = val.split("-")
            return (float(lo) + float(hi)) / 2
        except:
            return None
    # Handle single numbers like "10"
    try:
        return float(val)
    except:
        return None

merged['longevity_num'] = merged['longevity__dim'].apply(parse_longevity)

# --- Now compute average longevity by breed ---
dogs_longevity = (
    merged.groupby('breed_dim')['longevity_num']
    .mean()
    .reset_index(name='avg_longevity')
)

display(dogs_longevity.head(10))
```

```python
# --- Clean longevity__dim into numeric values and replace it ---
def parse_longevity(val):
    if pd.isna(val):
        return None
    val = str(val).strip()
    # Handle ranges like "12-15"
    if "-" in val:
        try:
            lo, hi = val.split("-")
            return (float(lo) + float(hi)) / 2
        except:
            return None
    # Handle single numbers like "10"
    try:
        return float(val)
    except:
        return None

# Replace old column with numeric values
merged['longevity__dim'] = merged['longevity__dim'].apply(parse_longevity)

# Quick check of the cleaned column
print(merged['longevity__dim'].head(15))

# Example: compute average longevity by breed
dogs_longevity = (
    merged.groupby('breed_dim')['longevity__dim']
    .mean()
    .reset_index(name='avg_longevity')
)

display(dogs_longevity.head(10))
```

```python
# --- 1. Clean weight__dim (convert ranges like "20-30" into average) ---
def parse_range(val):
    if pd.isna(val):
        return None
    val = str(val).strip()
    if "-" in val:
        try:
            lo, hi = val.split("-")
            return (float(lo) + float(hi)) / 2
        except:
            return None
    try:
        return float(val)
    except:
        return None

if 'weight__dim' in merged.columns:
    merged['weight__dim'] = merged['weight__dim'].apply(parse_range)

# --- 2. Clean height__dim (same logic as weight) ---
if 'height__dim' in merged.columns:
    merged['height__dim'] = merged['height__dim'].apply(parse_range)

# --- 3. Clean tradeamount in pet_supplies (extract sold count) ---
def parse_tradeamount(val):
    if pd.isna(val):
        return None
    val = str(val)
    import re
    numbers = re.findall(r'\d+', val)
    if numbers:
        return int(numbers[0])
    return None

if 'tradeamount' in supplies.columns:
    supplies['tradeamount_num'] = supplies['tradeamount'].apply(parse_tradeamount)

# --- 4. Clean price (remove symbols, convert to float) ---
def parse_price(val):
    if pd.isna(val):
        return None
    val = str(val).replace('$','').replace(',','').strip()
    try:
        return float(val)
    except:
        return None

if 'price' in supplies.columns:
    supplies['price_num'] = supplies['price'].apply(parse_price)

# --- Quick checks ---
print("Weight sample:", merged['weight__dim'].head(10).tolist() if 'weight__dim' in merged.columns else "No weight column")
print("Height sample:", merged['height__dim'].head(10).tolist() if 'height__dim' in merged.columns else "No height column")

if 'tradeamount_num' in supplies.columns:
    print(supplies[['tradeamount','tradeamount_num']].head(10))

if 'price_num' in supplies.columns:
    print(supplies[['price','price_num']].head(10))
```

```python
# --- 1. Longevity ---
if 'longevity__dim' in merged.columns:
    merged['longevity'] = merged['longevity__dim']  # already numeric after cleaning
    merged.drop(columns=['longevity__dim','longevity__desc'], inplace=True, errors='ignore')

# --- 2. Weight ---
if 'weight__dim' in merged.columns:
    merged['weight'] = merged['weight__dim']
    merged.drop(columns=['weight__dim'], inplace=True, errors='ignore')

# --- 3. Height ---
if 'height__dim' in merged.columns:
    merged['height'] = merged['height__dim']
    merged.drop(columns=['height__dim','height__desc'], inplace=True, errors='ignore')

# --- 4. Breed and Size Category ---
if 'breed_dim' in merged.columns:
    merged.rename(columns={'breed_dim':'breed'}, inplace=True)

if 'size_category_dim' in merged.columns:
    merged.rename(columns={'size_category_dim':'size_category'}, inplace=True)

# --- 5. Pet Supplies cleaning ---
if 'tradeamount_num' in supplies.columns:
    supplies.rename(columns={'tradeamount_num':'tradeamount_clean'}, inplace=True)

if 'price_num' in supplies.columns:
    supplies.rename(columns={'price_num':'price_clean'}, inplace=True)

# --- Quick check of cleaned tables ---
print("Merged columns:", merged.columns.tolist()[:20])  # show first 20 columns
print("Supplies columns:", supplies.columns.tolist())
```
